'''Q4: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)'''

'''A box plot, also known as a box-and-whisker plot, is a statistical visualization tool that displays the distribution of a dataset. It provides a summary of the key features of the distribution, such as the median, quartiles, and potential outliers. Box plots are particularly useful for comparing the distribution of multiple datasets and identifying the spread and central tendency of the data.

Here's why box plots are used:

Central Tendency and Spread: Box plots show the median, quartiles, and the range of the data, giving a sense of the central tendency and spread of the distribution.

Outlier Detection: Box plots include "whiskers" that extend to indicate the range of typical data values, and individual points beyond the whiskers can be identified as potential outliers.

Comparison of Distributions: Box plots make it easy to compare the distribution of multiple datasets, making them useful for identifying patterns, variations, and differences between groups.'''

'''Now, let's create a box plot using the provided data:'''

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Generate data
box1 = np.random.normal(100, 10, 200)
```

```python
box2 = np.random.normal(90, 20, 200)
```

```python
# Combine data into a single list
```

```python
data = [box1, box2]
```

```python
# Create a box plot
```

```python
plt.figure(figsize=(8, 6))
```

```python
plt.boxplot(data, labels=['Box 1', 'Box 2'])
```

```python
plt.title('Box Plot of Box 1 and Box 2')
```

```python
plt.ylabel('Values')
```

```python
# Show the plot
```

```python
plt.show()
```

'''The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to display multiple plots in the same figure, making it easier to compare and analyze different datasets or visualizations.

Here's an example of how to use subplot() to draw four line plots with the provided data:'''

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Data for four lines
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
```

```python
y1 = np.array([0, 100, 200, 300, 400, 500])
```

```python
y2 = np.array([50, 20, 40, 20, 60, 70])
```

```python
y3 = np.array([10, 20, 30, 40, 50, 60])
```

```python
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
# Create subplots
```

```python
plt.figure(figsize=(10, 8))
```

```python
# Subplot 1
plt.subplot(2, 2, 1)
```

```python
plt.plot(x, y1, label='Line 1')
```

```python
plt.title('Line Plot 1')
```

```python
# Subplot 2
```

```python
plt.subplot(2, 2, 2)
```

```python
plt.plot(x, y2, label='Line 2')
```

```python
plt.title('Line Plot 2')
```

```python
# Subplot 3
```

```python
plt.subplot(2, 2, 3)
```

```python
plt.plot(x, y3, label='Line 3')
```

```python
plt.title('Line Plot 3')
```

```python
# Subplot 4
```

```python
plt.subplot(2, 2, 4)
```

```python
plt.plot(x, y4, label='Line 4')
```

```python
plt.title('Line Plot 4')
```

```python
# Adjust layout to prevent overlap
```

```python
plt.tight_layout()
```

```python
# Show the plot
```

```python
plt.show()
```
