```python
from matplotlib import pyplot as plt
```

```python
from matplotlib import style as st
```

```python
days=[1,2,3,4,5,6,7,8,9]
```

```python
delhi_tem = [36,37,37.7,39,40.1,43,43,38,45]
mumbai_tem= [39,39.4,40,40.7,41,42.5,43.5,44,44.9]
```

```python
st.use('ggplot')
```

```python
plt.plot(days, delhi_tem, "r--", linewidth = 3,markersize= 10, label = "Delhi line")
plt.plot(days,mumbai_tem,"bo:",linewidth=3,
        markersize = 10, label = "Mumbai line")
plt.title("Goa vs Delhi", fontsize=15)
plt.xlabel("days",fontsize=13)
plt.ylabel("temp")
plt.axis([0,10,25,50])
plt.legend(loc = 1)
plt.grid(color='g', linestyle='-',linewidth=1)
plt.show()
```

```python
plt.plot(days,mumbai_tem,"bo:",linewidth=3,
        markersize = 10, label = "Temp line")
```

```python
cv=[25,35,24,16]
t=['Tata', 'Maruti', 'Hyundai', 'ford']
plt.pie(cv,labels=t,explode=(0,0,0.1,0), autopct='%.2f', shadow=True)
plt.show()
```
