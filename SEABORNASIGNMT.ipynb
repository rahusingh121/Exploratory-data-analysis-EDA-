'''Que 6: Use the "flights" dataset from seaborn to plot a heatmap.'''

'''The "flights" dataset is not available directly in the seaborn library, but it is present in the seaborn library's example datasets. You can load the "flights" dataset and then create a heatmap to visualize the data. Here's an example:'''

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
# Load the flights dataset from seaborn example datasets
```

```python
flights_data = sns.load_dataset("flights")
```

```python
# Pivot the data to create a matrix suitable for a heatmap
```

```python
flights_matrix = flights_data.pivot_table(index='month', columns='year', values='passengers')
```

```python
# Create a heatmap
```

```python
plt.figure(figsize=(12, 8))
```

```python
sns.heatmap(flights_matrix, annot=True, fmt='d', cmap='coolwarm')
```

```python
# Set plot labels and title
```

```python
plt.xlabel('Year')
```

```python
plt.ylabel('Month')
```

```python
plt.title('Heatmap of Flights Dataset')
```

```python
# Show the plot
```

```python
plt.show()
```

'''In this example:

pivot_table is used to reshape the data into a matrix suitable for a heatmap.
annot=True adds numerical annotations to each cell.
fmt='d' specifies the format of the annotations as integers.
cmap='coolwarm' sets the color map for the heatmap.
Adjust the plot parameters according to your preferences. This code will create a heatmap to visualize the number of passengers across months and years in the "flights" dataset.





'''
